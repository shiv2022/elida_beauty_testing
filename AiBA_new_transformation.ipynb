```python
import numpy as np
import pandas as pd
import json
import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.expand_frame_repr', False)
```

```python
# Loading the csv file 
df1 = pd.read_csv('/content/elida_beauty_testing/elida_data.csv')
df1.columns
```

```python
cols = df1.columns
df1.rename (columns = {cols[0]:'Q1',cols[1]:'Q2',cols[2]:'Q3',cols[3]:'Q4'}, inplace=True)

df1.columns
```

```python
# Splitting the values present in the column into multiple values based on the separator	
df1 = df1.assign(Q1=df1.Q1.str.split(", ")).explode('Q1')
df1 = df1.assign(Q2=df1.Q2.str.split("/ ")).explode('Q2')
df1 = df1.assign(Q3=df1.Q3.str.split(",")).explode('Q3')
df1 = df1.assign(Q4=df1.Q4.str.split("/ ")).explode('Q4')
```

```python
df1.head()
```

```python
len(df1)
```

```python
# Removing the leading and trailing spaces from columns

# col_list = ['Q1','Q2','Q3','Q4']
for col in df1.columns:
  print(df1[col].unique())
  df1[col] = df1[col].str.strip()
  print(df1[col].unique())
```

```python
df6 = pd.merge(df5, df3, left_on='Q2', right_on='answerInEnglish')
df6.rename(columns={'questionId':'questionId2', 'questionText':'questionText2', 'questionDescription':'questionDescription2', 'multiSelect':'multiSelect2', 'answerInEnglish':'answerInEnglish2', 'responseID':'responseID2', 'responseDescription':'responseDescription2', 'responseText':'responseText2', 'seedWords':'seedWords2'}, inplace=True)


df7 = pd.merge(df6, df3, left_on='Q3', right_on='answerInEnglish')
df7.rename(columns={'questionId':'questionId3', 'questionText':'questionText3', 'questionDescription':'questionDescription3', 'multiSelect':'multiSelect3', 'answerInEnglish':'answerInEnglish3', 'responseID':'responseID3', 'responseDescription':'responseDescription3', 'responseText':'responseText3', 'seedWords':'seedWords3'}, inplace=True)

df8 = pd.merge(df7, df3, left_on='Q4', right_on='answerInEnglish')
df8.rename(columns={'questionId':'questionId4', 'questionText':'questionText4', 'questionDescription':'questionDescription4', 'multiSelect':'multiSelect4', 'answerInEnglish':'answerInEnglish4', 'responseID':'responseID4', 'responseDescription':'responseDescription4', 'responseText':'responseText4', 'seedWords':'seedWords4'}, inplace=True)
```

```python
# Working with pim_df sheet

pim_df = pd.read_csv('/content/elida_beauty_testing/Updated_Pim_sheet.csv')
pim_df.columns
```

```python
pim_df = pim_df.dropna(how='all', axis=1)
pim_df.columns
```

```python
# df9 = pim_df[['productName','identifierValue']].copy()
df9 = pim_df.copy()
# df9.columns
df9['identifierValue'] = df9['identifierValue'].str.strip('GTIN:')

# Stripping any unwanted space before and after the values in the columns
for col in df9.columns:
  df9[col] = df9[col].str.strip()
```

```python
df9.columns
```

```python
df9.head()
```

```python
df8_new = df8.copy()

df10 = pd.merge(df8_new, df9, left_on='P1', right_on='productName');
df10.rename(columns={'identifierValue':'GTIN1','category':'category1','productDescription':'productDescription1','pdpUrl':'pdpUrl1','imageUrl':'imageUrl1'},inplace=True)
df10.drop(['productName'], axis=1, inplace=True)

df11 = pd.merge(df10, df9, left_on='P2', right_on='productName')
df11.rename(columns={'identifierValue':'GTIN2','category':'category2','productDescription':'productDescription2','pdpUrl':'pdpUrl2','imageUrl':'imageUrl2'},inplace=True)
df11.drop(['productName'], axis=1, inplace=True)

df12 = pd.merge(df11, df9, left_on='P3', right_on='productName')
df12.rename(columns={'identifierValue':'GTIN3','category':'category3','productDescription':'productDescription3','pdpUrl':'pdpUrl3','imageUrl':'imageUrl3'},inplace=True)
df12.drop(['productName'], axis=1, inplace=True)

df13 = pd.merge(df12, df9, left_on='P4', right_on='productName')
df13.rename(columns={'identifierValue':'GTIN4','category':'category4','productDescription':'productDescription4','pdpUrl':'pdpUrl4','imageUrl':'imageUrl4'},inplace=True)
df13.drop(['productName'], axis=1, inplace=True)
```

```python
# Creating a composite key to save the individual json files

df16 = df15[['responseID1','responseID2','responseID3','responseID4','GTIN1']].apply(lambda x: '_'.join(x.astype(str)), axis=1)
```

```python
for composite_key in df16:
  print(composite_key)
```

```python
# iterate through each row of the dataframe
for index, row in df15.iterrows():

  # convert the row to a dictionary
  data = row.to_dict()
  for composite_key in df16:
    # write the dictionary to a json file
    with open(f'{composite_key}.json', 'w') as f:
      json.dump(data, f)
```
